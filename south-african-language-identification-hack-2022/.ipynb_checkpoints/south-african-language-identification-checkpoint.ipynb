**Word Cloud of Each Language ID**
* We can collect word via our Word Cloud function that takes two arguments: dataframe and specific lang_id (lang_id as a string)

```python
def word_cloud(df, label):
    # extract keywords according to news sentiments==2
    df_keywords = df['text'][df['lang_id']==label]
    new_df = pd.DataFrame(df_keywords)

    word_vect = CountVectorizer(stop_words = 'english')
    keywords = word_vect.fit_transform(new_df.text)
    total_words = keywords.sum(axis=0)

    keywords_freq = [(keyword, total_words[0, i]) for keyword, i in word_vect.vocabulary_.items()]
    keywords_freq = sorted(keywords_freq, key = lambda x: x[1], reverse = True)
    frequency = pd.DataFrame(keywords_freq, columns=['keyword', 'freq'])

    kw_cloud = WordCloud(width=600, 
                     height=400, 
                     random_state=2, 
                     max_font_size=100).generate(str(keywords_freq))
    
    x, y = np.ogrid[:300, :300]
    mask = (x - 160) ** 2 + (y - 160) ** 2 > 170 ** 2
    mask = 255 * mask.astype(int)

    plt.figure(figsize=(10,8))
    b = plt.imshow(kw_cloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(label, fontsize = 20)

    return b
```

**Preprocessing**
* Cleaning urls and all website links using regular expressions
* Keeping relevant Alphabetical and Numerical Characters
* Tokenize and Lemmatize text
* Remove Stopwords
* Re-join the text words
* Create a bag of words for our model

```python
nltk.download('omw-1.4')

# create a function to do some preprocessing on the train dataset
cleanedData = []

lemma = WordNetLemmatizer()
stop_words = stopwords.words("english")
for wds in df_train["text"]:
    
    # Cleaning links
    wds = re.sub(r'http\S+', '', wds)
    
    # Cleaning everything except alphabetical and numerical characters
    wds = re.sub("[^a-zA-Z0-9]"," ",wds)
    
    # Tokenizing and lemmatizing
    wds = nltk.word_tokenize(wds.lower())
    wds = [lemma.lemmatize(word) for word in wds]
    
    # Removing stopwords
    wds = [word for word in wds if word not in stop_words]
    
    # Joining
    wds = " ".join(wds)
    
    cleanedData.append(wds)
```

```python
# create a function to do some preprocessing on the test dataset
cleanedData_test = []

lemma = WordNetLemmatizer()
stop_words = stopwords.words("english")
for wds in df_test["text"]:
    
    # Cleaning links
    wds = re.sub(r'http\S+', '', wds)
    
    # Cleaning everything except alphabetical and numerical characters
    wds = re.sub("[^a-zA-Z0-9]"," ",wds)
    
    # Tokenizing and lemmatizing
    wds = nltk.word_tokenize(wds.lower())
    wds = [lemma.lemmatize(word) for word in wds]
    
    # Removing stopwords
    wds = [word for word in wds if word not in stop_words]
    
    # Joining
    wds = " ".join(wds)
    
    cleanedData_test.append(wds)
```

**Vectorizing, Standardizing and Splitting Dataset**

```python
# this function takes our cleaned dataset as argument,
# standardizes and transforms it for our model
def preprocess_train_split(df):
    count_vect = CountVectorizer(max_features = 2500)
    
    # split dataset into features and label
    # then convert feature to numpy array
    features = count_vect.fit_transform(cleanedData).toarray()
    label = df['text']
    
    # instantiate the MinMaxScaler() and transform features
    # split train data into train and validation datasets
    features_stand = MinMaxScaler().fit_transform(features)
    
    X_train, X_val, y_train, y_val = train_test_split(features_stand, 
                                                    label, 
                                                    test_size=0.20, 
                                                    random_state=42)
    
    return (X_train, y_train), (X_val, y_val)
```

```python
from sklearn.model_selection import train_test_split
(X_train, y_train), (X_val, y_val) = preprocess_train_split(df_train)

print(X_train.shape)
print(X_val.shape)
print(y_train.shape)
print(y_val.shape)
```

```python
# function to standardize test dataset
def preprocess_test(df):

    ''' returns standardized predictors of the test set using the MinMaxScaler class
    parameters:
        df: a dataframe containing all the features from the test set
    ''' 
    count_vect = CountVectorizer(max_features = 2500)
    features = count_vect.fit_transform(cleanedData_test).toarray()
    # instantiate the MinMaxScaler()
    # transform features of the test datasets
    features_stand = MinMaxScaler().fit_transform(features)
    X_test = features_stand
    
    return X_test
```

```python
X_test = preprocess_test(df_test)
print(X_test.shape)
```

**Support Vector Machine 'rbf'**

```python
svm = SVC(kernel='rbf', decision_function_shape='ovr',
          probability=True,degree=3, gamma=0.9, C=0.9, random_state=1)
```

*I borrowed the https://github.com/g-leibovitz/2207ACDS_TEAMND2 notebook functions for the train and test models*

```python
# create a function to train our models
def train_model(model, X, y):

    ''' returns a model trained on the training dataset
        parameters:
            model:   a machine learning model
            X: a training set of independent features - X_train or X_train_sm
            y: values of the target variable for training - X_train or X_train_sm
    '''    
    return model.fit(X, y)
```

```python
# Create a function to make predictions
def test_model(model, X):
    
    ''' returns the predictions of a model on the X_validation set 
        after predicting the values of y validation.
        parameters:
            model:  a machine learning model
            X:      validation set of features
    '''          
    return model.predict(X)
```

**Model Fitting, Training and Predictions of the Text Classes with unseen datasets**

```python
# training the support vector machine on balanced data
svm = train_model(svm, X_train, y_train)
```

```python
# predict for the support vector machine on balanced data
svm_pred_val = test_model(svm, X_val)
```

```python
# Call the function to give model predictions on test dataset
svm_pred_test = test_model(svm, X_test)
```

```python
# get model accuracy
print("Training Accuracy :", svm.score(X_train, y_train))
print("Validation Accuracy :", svm.score(X_val, y_val))

# confusion matrix
c_matrix = confusion_matrix(y_val, svm_pred_val)
print(c_matrix)
```

```python
#create a submission file for the model
results_dict = pd.DataFrame({'Text':df_test['text'],
                'Predicted lang_id': svm_pred_test})

results_dict.to_csv('submission.csv', index = False)
```
